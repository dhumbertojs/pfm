```python
import pandas as pd
from ckanapi import RemoteCKAN

# URL de la API de CKAN
url = 'https://datosabiertos.infocdmx.org.mx'

# Crear una instancia de RemoteCKAN
ckan = RemoteCKAN(url)

# Lista de resource_id predefinidos
resource_ids = ['e0d70ebb-7c16-46c7-8865-c537df8ff637', '2eb8f23e-a098-4746-80ce-e52562bd1f11', 
                'a27a3b30-b727-4650-896d-1258729e13ba', '132a3785-c718-4122-96df-031437d2f289',
               '5043bd4d-07bf-40db-a9df-8f2755aa36bc', 'de055dd6-2eb2-4d75-adb9-2041b47fdb4e',
               '2bef2739-b570-4c80-afb2-e8ccb59e8f2c', '52c81f47-214b-4888-9869-0b17b781ebb9',
               'd68925b2-fcd3-4676-b2c5-d73c3ac69d9a', '540d06f9-b9a2-4f56-b23d-1e84e2ed85a2',
               '5c68a7a5-be2d-4a4f-ad04-ff04a9ed2889', 'e8b2d6d6-c4cf-4a49-ac2b-512ab3597283',
               '074b6c4e-89fd-4119-a43a-69b1e007e1b3', 'd7019dab-918c-41e9-ad6d-9fb9838c5021',
               '2a29d720-32bc-4268-90da-3b374aaa4379', '43f4549c-9cba-4737-b86b-f2ee4924ddf8'#,
               #'9c683caf-f559-462c-8e70-845c8902d7b4'
                # NotFound: Resource "9c683caf-f559-462c-8e70-845c8902d7b4" was not found.
               ]  # Lista de IDs de recursos

# Lista para almacenar todos los registros
all_records = []

# Iterar sobre los resource_id predefinidos
for resource_id in resource_ids:
    # Parámetros de la consulta
    limit = 1000  # Número máximo de registros por consulta
    offset = 0  # Desplazamiento inicial de los registros

    # Realizar la consulta a la API en lotes hasta obtener todos los registros
    while True:
        # Parámetros de la consulta actual
        params = {
            'resource_id': resource_id,
            'limit': limit,
            'offset': offset
        }

        # Realizar la consulta a la API
        data = ckan.action.datastore_search(**params)

        # Obtener los registros de la respuesta actual
        records = data['records']

        # Agregar los registros a la lista general
        all_records.extend(records)

        # Actualizar el desplazamiento para la próxima consulta
        offset += limit

        # Verificar si se han obtenido todos los registros
        if len(records) < limit:
            break

# Crear un DataFrame de Pandas con todos los registros
df = pd.DataFrame(all_records)

# Imprimir el DataFrame resultante
print(df)
```
